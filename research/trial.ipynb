```python
import sys
import os

project_root = os.path.abspath(os.path.join(os.getcwd(), '..')) 
if project_root not in sys.path:
    sys.path.append(project_root)

from backend.src.agents.assistant_agent import AssistantAgent
from backend.src.agents.tutoring_agent import TutoringAgent
from backend.src.agents.testing_agent import TestingAgent

topic = "Calculus"
subject = "Mathematics"
standard = 11

assistant_agent = AssistantAgent()

assistant_response = await assistant_agent.run(
    topic=topic,
    subject=subject,
    standard=standard
)

tutoring_agent = TutoringAgent()

tutor_response = await tutoring_agent.run(
    instructions = assistant_response,
    standard = standard,
    subject = subject,
    topic = topic
)

testing_agent = TestingAgent()

test_response = await testing_agent.run(tutor_response)
```

```python
for title, content in tutor_response.items():

    print(title, "\n")
    print(content)
    print("\n")
    print("-" * 100)    
```

```python
for title, test in test_response.items():

    print(title, "\n")
    print(test)
    print("\n")
    print("-" * 100)
```
